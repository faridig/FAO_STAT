```python
import pandas as pd
import numpy as np
```

```python
df_anim = pd.read_csv("animal.csv")
df_veg = pd.read_csv("vegetal.csv")
df_pop = pd.read_csv("population.csv")
```

<font color='green'>**1 .Nettoyer les titres de colonnes :
Supprimer les espaces au début et à la fin des titres
Remplacer les espaces par des underscores (ceux se situant entre les mots)
Tout mettre en minuscule**</font>


```python
# Converti les noms de colonnes en chaînes de caractères
df_anim.columns = df_anim.columns.astype(str)


df_anim.columns = df_anim.columns.str.strip()
df_anim.columns = df_anim.columns.str.replace(" ", "_")
df_anim.columns = df_anim.columns.str.lower()

```

```python
# Converti les noms de colonnes en chaînes de caractères
df_veg.columns = df_veg.columns.astype(str)


df_veg.columns = df_veg.columns.str.strip()
df_veg.columns = df_veg.columns.str.replace(" ", "_")
df_veg.columns = df_veg.columns.str.lower()
```

```python
# Converti les noms de colonnes en chaînes de caractères
df_pop.columns = df_pop.columns.astype(str)


df_pop.columns = df_pop.columns.str.strip()
df_pop.columns = df_pop.columns.str.replace(" ", "_")
df_pop.columns = df_pop.columns.str.lower()
```
