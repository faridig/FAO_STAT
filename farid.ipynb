```python
import pandas as pd
df_pop = pd.read_csv('population.csv') 
df_veg = pd.read_csv('vegetal.csv')
df_anim = pd.read_csv('animal.csv')

df_pop.shape
df_veg.shape
df_anim.shape
```

```python
# 1) nettoyage des colonnes
df_anim.columns = df_anim.columns.str.strip()
df_anim.columns = df_anim.columns.str.lower()
df_anim.rename(columns=lambda x: x.replace(' ','_'), inplace=True)

df_veg.columns = df_veg.columns.str.strip()
df_veg.columns = df_veg.columns.str.lower()
df_veg.rename(columns=lambda x: x.replace(' ','_'), inplace=True)

df_pop.columns = df_pop.columns.str.strip()
df_pop.columns = df_pop.columns.str.lower()
df_pop.rename(columns=lambda x: x.replace(' ','_'), inplace=True)

```

```python
# 4 
df_veg["type"]='vegetal'
df_anim['type'] = 'animal'
```

```python
# question 4
df_product = pd.concat([df_anim, df_veg], axis=0)
df_product.head()

```

```python
# question 5 : nettoyage colonne population et renommer colonne 
df_pop = df_pop.drop(['domain_code','domain','element_code','element','item_code_(fbs)','item','year_code','unit','flag','flag_description'], axis=1)
df_pop = df_pop.rename(columns={'value':'pop_1000_hab'})
df_pop.head()
```

```python
# 6. Transformez product afin de ne garder que les colonnes area_code_(fao), area, element, item, year, type, unit.

df_product = df_product.drop(['domain_code', 'domain', 'element_code', 'item_code_(fbs)','year_code','flag', 'flag_description'], axis=1)


df_product.head()
```

```python
# 7 Fusionnez df_pop avec product et nommez ce DataFrame df.

df = pd.merge(df_product,df_pop, how='left', on=['area_code_(m49)','area','year'])
df = df.rename(columns={'area_code_(m49)':'country_code','area':'country'})

df.head()

```

```python
df['pop_1000_hab']= df['pop_1000_hab']*1000

def formater_nombre(nombre):
    return '{:,.0f}'.format(nombre).replace(',', ' ')

df['nombre_formate'] = df['pop_1000_hab'].apply(formater_nombre)


df.head()
```

```python


# df = df.drop('pop_1000_hab', axis=1)
df.rename(columns={'nombre_formate':'population'}, inplace=True)

df.head()
```

```python
# 11.Affichez les valeurs uniques de la colonne country

df['country'].unique()
```

```python
# Modifiez votre jeu de données afin que les informations soient indexées par country_code, country, year, pop_1000_hab, type et item. Les valeurs de la colonne element doivent être séparées dans des colonnes différentes. Recherchez sur internet la différence entre les formats long et les formats wide. Pour réussir cette étape creuser la méthode pivot_table. Il est préférable d’appliquer la méthode reset_index() après avoir utilisé la méthode précédente.

df_wide = df.pivot_table(index=['country_code', 'country', 'year', 'population', 'type','item'], columns='element', values='value').reset_index()
df_wide.head()
```

```python
# 14. Faire du nettoyage dans le nom des colonnes
# Notez bien la dimension du jeu de données. Si tout s'est bien déroulé vous devez avoir le même resultat, sinon revoyez les étapes précédentes.

df_wide = df_wide.rename(columns={'item' : 'product'})
df_wide = df_wide.rename(columns={'type' : 'product_category'})
df_wide.columns = df_wide.columns.str.strip()
df_wide.columns = df_wide.columns.str.lower()
```
