```python
# 16. Nous allons ajouter une nouvelle colonne à notre jeu de données : la zone géographique. 
# Pour cela rendez-vous sur le site de la FAO, votre objectif est de télécharger des nouveaux datasets pour ajouter les zones suivantes à votre jeu de données : 'eastern_africa', 'middle_africa', 'northern_africa', 'southern_africa', 'western_africa', 'northern_america', 'central_america', 'caribbean', 'south_america', 'central_asia', 'eastern_asia', 'southern_asia', 'south_eastern_asia', 'western_asia', 'eastern_europe', 'northern_europe', 'southern_europe', 'western_europe', 'australia_new_zealand', 'melanesia', 'micronesia', 'polynesia'

import os

def process_csv(file_path):
    df = pd.read_csv(file_path)
    nom_sans_extension = os.path.basename(file_path).split('.')[0]
    df['geographical_area'] = nom_sans_extension
    colonnes = ['Area', 'geographical_area']
    new_df = df[colonnes]
    return new_df


csv_files = [
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/australia_new_zealand.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/caribbean.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/central_america.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/central_asia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/eastern_africa.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/eastern_asia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/eastern_europe.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/melanesia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/micronesia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/middle_africa.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/northern_africa.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/northern_america.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/northern_europe.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/polynesia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/south_america.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/south_eastern_asia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/southern_africa.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/southern_asia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/southern_europe.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/western_africa.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/western_asia.csv',
    '/home/utilisateur/Documents/FAO_STAT/zone_geo/western_europe.csv'
]


results = []

for csv_file in csv_files:
    result = process_csv(csv_file)
    results.append(result)

df_geo = pd.concat(results)
df_geo.drop_duplicates(inplace = True)
df_geo
```
